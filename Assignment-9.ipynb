```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=df, palette="Set1")
plt.title("Total Bill vs Tip separated by Sex")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.stripplot(x="day", y="tip", data=df, jitter=True, hue="day", palette="pastel", legend=False)
plt.title("Scatter of Tips across Days")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.stripplot(x="time", y="total_bill", data=df, hue="time", palette="Set2", jitter=True, legend=False)
plt.title("Scatter of Total Bill across Meal Times")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="total_bill", y="tip", hue="smoker", style="time", data=df, palette="Dark2")
plt.title("Total Bill vs Tip separated by Smoker & Time")
plt.show()
```
