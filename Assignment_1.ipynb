```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins")
df = df.drop(columns=["body_mass_g"])
df = df.dropna()
num_df = df.select_dtypes(include="number")
```

```python
summary_stats = num_df.describe().T
summary_stats["range"] = summary_stats["max"] - summary_stats["min"]
summary_stats["variance"] = num_df.var()
summary_stats["std_dev"] = num_df.std()
print("\nSummary Statistics:\n")
print(summary_stats)
```

```python
plt.figure(figsize=(8,5))
num_df.mean().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Mean of Each Feature (Penguins Dataset)")
plt.ylabel("Mean Value")
plt.show()
```

```python
num_df.hist(bins=15, figsize=(10,6), color="lightgreen", edgecolor="black")
plt.suptitle("Histograms of Penguin Features")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=num_df, palette="Set2")
plt.title("Boxplots of Penguin Features")
plt.xticks(rotation=45)
plt.show()
```
