```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
cat_cols = ["sex", "class", "embarked", "who", "deck"]
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,8))  # 2 rows, 3 columns
axes = axes.flatten()
for i, col in enumerate(cat_cols):
    counts = df[col].value_counts(dropna=False)
    counts.plot(kind='bar',
                ax=axes[i],
                color=sns.color_palette("pastel", n_colors=len(counts)))
    axes[i].set_title(f"Frequency of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")
for j in range(len(cat_cols), len(axes)):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,8))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    proportions = df[col].value_counts(normalize=True, dropna=False)
    proportions.plot.pie(
        autopct="%1.1f%%",
        colors=sns.color_palette("pastel", n_colors=len(proportions)),
        ax=axes[i]
    )
    axes[i].set_ylabel("")
    axes[i].set_title(f"Proportion of {col}")

for j in range(len(cat_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,8))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(x=col, hue="survived", data=df, palette="Set2", ax=axes[i])
    axes[i].set_title(f"{col} vs Survived")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")

for j in range(len(cat_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
